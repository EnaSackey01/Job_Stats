```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
from scipy.stats import kstest, norm

Cleaned_Job_df = pd.read_csv("Joined_Data/Cleaned_New_Joined_Data.csv")

Cleaned_Job_df['salary_standardized']
```

```python
Cleaned_Job_df.shape
```

```python
# All data analyst salaries

sns.histplot(data = Cleaned_Job_df, x = 'salary_standardized', kde = True).set(title = 'Full Salaries of All Data Analysts')

plt.savefig("Images/full_salary_hist.jpg")
```

```python
# Histograms salary_standardized

salary_standardized = Cleaned_Job_df["salary_standardized"]

salary_standardized.hist()

plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Histogram of Standard Salaries")

plt.savefig("Images/Histogram of Standard Salaries.jpg")
```
